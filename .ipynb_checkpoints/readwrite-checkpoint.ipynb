```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
rawDfrw = pd.read_csv('sar_mpgu_izh.csv', skiprows=57064, sep='\s+', skipinitialspace=True, nrows=(61355-57065), engine='python')
rawDfrw.tail()
```

```python
time = rawDfrw.iloc[:,0]
time = pd.to_datetime(time).apply(lambda t: t.strftime('%H:%M'))
y1 = rawDfrw.iloc[:, 7].astype(float)
y2 = rawDfrw.iloc[:, 8].astype(float)
print(y1)

```

```python
dfAw = pd.DataFrame({'Time': time, 'DEV': rawDfrw.iloc[:,1], 'await': y1})
dfSv = pd.DataFrame({'Time': time, 'DEV': rawDfrw.iloc[:,1], 'svctm': y2})
# dfSv.tail()
dfAw_temp = dfAw[dfAw['await'] != 0]

dev_names = set(dfAw_temp['DEV'])
print(dev_names)
# dfAw = dfAw[dfAw['DEV'] == 'dev253-3']
# dfAw.tail()
```

```python
dfAw1 = dfAw[dfAw['DEV'] == 'dev253-4']
dfAw2 = dfAw[dfAw['DEV'] == 'dev253-3']
dfAw3 = dfAw[dfAw['DEV'] == 'dev253-6']
dfAw4 = dfAw[dfAw['DEV'] == 'dev253-0']
dfAw5 = dfAw[dfAw['DEV'] == 'dev8-0']
dfAw6 = dfAw[dfAw['DEV'] == 'dev8-16']
# dfAw2.tail()
```

```python
dfSv1 = dfSv[dfSv['DEV'] == 'dev253-4']
dfSv2 = dfSv[dfSv['DEV'] == 'dev253-3']
dfSv3 = dfSv[dfSv['DEV'] == 'dev253-6']
dfSv4 = dfSv[dfSv['DEV'] == 'dev253-0']
dfSv5 = dfSv[dfSv['DEV'] == 'dev8-0']
dfSv6 = dfSv[dfSv['DEV'] == 'dev8-16']
# dfSv6.tail()
```

```python
dfAw1.reset_index(inplace=True, drop=True)
dfAw2.reset_index(inplace=True, drop=True)
dfAw3.reset_index(inplace=True, drop=True)
dfAw4.reset_index(inplace=True, drop=True)
dfAw5.reset_index(inplace=True, drop=True)
dfAw6.reset_index(inplace=True, drop=True)

dfSv1.reset_index(inplace=True, drop=True)
dfSv2.reset_index(inplace=True, drop=True)
dfSv3.reset_index(inplace=True, drop=True)
dfSv4.reset_index(inplace=True, drop=True)
dfSv5.reset_index(inplace=True, drop=True)
dfSv6.reset_index(inplace=True, drop=True)
```

```python
# График для dev253-4
fig, ax = plt.subplots(figsize=(12.992, 6.496))
ax.plot(dfAw1.Time, dfAw1['await'], 'b',label='Среднее время выполнения чтения/записи')
ax.plot(dfSv1.Time, dfSv1['svctm'], 'g', label='Среднее время обслуживания чтения/записи')
plt.axvline(x='11:46', color='r')
ax.set_xlabel('Продолжительность теста')
ax.set_ylabel('Время, мс', rotation=0, y=1.05)
ax.set_title('Среднее время чтения/записи (dev253-4)', y=1.05)
ax.legend(loc=1)
ax.tick_params(axis='x', labelrotation=90)
# ax.set_ylim(0,20)

xAxis = np.arange(0, dfAw1.index[-1],5)
plt.xlim(dfAw1['Time'][0], dfAw1['Time'][dfAw1.index[-1]])
plt.xticks(xAxis, dfAw1['Time'][xAxis], rotation='vertical')

plt.savefig('Среднее время чтениязаписи_1')
plt.show()
plt.close()
```
