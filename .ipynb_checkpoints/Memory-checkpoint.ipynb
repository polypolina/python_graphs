```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
rawDfMem = pd.read_csv('sar_mpgu_izh.csv', skiprows=56272, sep='\s+', skipinitialspace=True, nrows=(56468-56273), engine='python')
rawDfMem.tail()
```

```python
time = rawDfMem.iloc[:,0]
time = pd.to_datetime(time).apply(lambda t: t.strftime('%H:%M'))
dfMem = pd.DataFrame({'Time': time, 'memused': rawDfMem.iloc[:,rawDfMem.columns.get_loc('%memused')]})
dfMem.tail()
```

```python
rawDfSwp = pd.read_csv('sar_mpgu_izh.csv', skiprows=56470, sep='\s+', skipinitialspace=True, nrows=(56666-56471), engine='python')
rawDfSwp.tail()
```

```python
time = rawDfSwp.iloc[:,0]
time = pd.to_datetime(time).apply(lambda t: t.strftime('%H:%M'))
dfSwp = pd.DataFrame({'Time':time, 'swpused': rawDfSwp.iloc[:,rawDfSwp.columns.get_loc('%swpused')]})
dfSwp.tail()
```

```python
# График утилизации памяти
fig, ax = plt.subplots(figsize=(12.992, 6.496))
ax.plot(dfMem.Time, dfMem['memused'], 'b',label='Утилизация памяти')
plt.axvline(x='11:46', color='r')
ax.set_xlabel('Продолжительность теста')
ax.set_ylabel('Утилизация памяти, %', rotation=0, y=1.05)
ax.set_title('Утилизация памяти', y=1.05)
ax.legend(loc=2)
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylim(60,70)

# График утилизации подкачки
ax2 = ax.twinx()
ax2.set_ylim(0.05, 0.2)
ax2.plot(dfSwp.Time, dfSwp['swpused'], 'g', label='Утилизация подкачки')
ax2.legend(loc=1)
ax2.xaxis.label.set_visible(False)
# ax2.xaxis.set_visible(False)
ax2.set_ylabel('Утилизация подкачки, %', rotation=0, y=1.05)

xAxis = np.arange(0, dfMem.index[-1],5)
plt.xlim(dfMem['Time'][0], dfMem['Time'][dfMem.index[-1]])
plt.xticks(xAxis, dfMem['Time'][xAxis], rotation='vertical')

plt.savefig('Утилизация памяти')
plt.show()
plt.close()
```
