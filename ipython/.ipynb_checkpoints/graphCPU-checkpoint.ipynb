```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
# rawDfCPU = pd.read_csv('sar_mpgu_izh.csv', skiprows=2, sep='\s+', skipinitialspace=True, header=None, nrows=4906, engine='python')
# rawDfCPU.tail()
```

```python
rawDfCPU = pd.read_csv('sar_mpgu_izh.csv', skiprows=2, sep='\s+', skipinitialspace=True, nrows=4879, engine='python')
rawDfCPU.tail()
```

```python
yVal = rawDfCPU.iloc[:, rawDfCPU.columns.get_loc('%idle')].replace('%idle', np.nan).astype(float)
print(yVal)
```

```python
dfCPU = pd.DataFrame({'Time':rawDfCPU.iloc[:,0], 'CPU':rawDfCPU.iloc[:,rawDfCPU.columns.get_loc('CPU')], 'dif': 100-yVal})
dfCPU = dfCPU[dfCPU['CPU']=='all']
# dfCPU.head()
dfCPU.reset_index(inplace=True, drop=True)
dfCPU.tail()
```

```python
#  CPU
fig, axes = plt.subplots(figsize=(10.629, 6.496))
axes.plot(dfCPU.Time, dfCPU.dif, 'b',label=' CPU')
axes.set_xlabel(' ')
axes.set_ylabel(' CPU, %', rotation=0, y=1.05)
axes.set_title(' CPU')
axes.legend(loc=2)

xAxis = np.arange(0, dfCPU.index[-1],5)
plt.xlim(dfCPU['Time'][0], dfCPU['Time'][dfCPU.index[-1]])
plt.xticks(xAxis, dfCPU['Time'][xAxis], rotation='vertical')
```

```python
rawDfQ = pd.read_csv('sar_mpgu_izh.csv', skiprows=56866, sep='\s+', skipinitialspace=True, nrows=195, engine='python')
rawDfQ.tail()
```

```python
dfQ = pd.DataFrame({'Time': rawDfQ.iloc[:,0], 'runq-sz': rawDfQ.iloc[:,1]})
dfQ.head()
```
